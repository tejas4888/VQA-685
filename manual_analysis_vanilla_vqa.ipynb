```python
test_predictions = []
with (open("test_sents_preds_dict-all-classes-img-and-text.pkl", "rb")) as openfile:
    test_predictions = pickle.load(openfile)
```

```python
len(test_predictions)
```

```python
wrong_predictions = []
for prediction in test_predictions:
  if (prediction['predicted_answer'] != next(iter(prediction['label']))):
    wrong_predictions.append(prediction)
```

```python
len(wrong_predictions)
```

```python
indices = random.sample(range(0, len(wrong_predictions) + 1), 150)
```

```python
manual_eval_predictions = [wrong_predictions[i] for i in indices]
```

```python
n = len(manual_eval_predictions)
```

```python
manual_eval_predictions[0]
```

```python
answer_type = {}
for prediction in manual_eval_predictions:
  if(prediction['answer_type'] in answer_type):
    answer_type[prediction['answer_type']] += 1
  else:
    answer_type[prediction['answer_type']] = 1
```

```python
question_type = {}
for prediction in manual_eval_predictions:
  if(prediction['question_type'] in question_type):
    question_type[prediction['question_type']] += 1
  else:
    question_type[prediction['question_type']] = 1
```

```python
predicted_class = {}
for prediction in manual_eval_predictions:
  if(prediction['predicted_class'] in predicted_class):
    predicted_class[prediction['predicted_class']] += 1
  else:
    predicted_class[prediction['predicted_class']] = 1
```

```python
predicted_answer = {}
for prediction in manual_eval_predictions:
  if(prediction['predicted_answer'] in predicted_answer):
    predicted_answer[prediction['predicted_answer']] += 1
  else:
    predicted_answer[prediction['predicted_answer']] = 1
```

```python
questions_what = []
question_is = []
for prediction in manual_eval_predictions:
  if(prediction['question_type'] == 'what'):
    # print(prediction)
    questions_what.append(prediction)
  if(prediction['question_type'] == 'is'):
    question_is.append(prediction)
```

```python
sent_types = {}
for prediction in manual_eval_predictions:
  if(prediction['sent'] in sent_types):
    sent_types[prediction['sent']] += 1
  else:
    sent_types[prediction['sent']] = 1
```
