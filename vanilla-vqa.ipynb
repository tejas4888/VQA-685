```python
from dataset import *
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.models import Model
import numpy as np
import pickle
```

```python
train_dict = get_data_dictionary("train")
test_dict = get_data_dictionary("test")
val_dict = get_data_dictionary("val")

data_dict={}
data_dict["train"]=train_dict
data_dict["test"]=test_dict
data_dict["val"]=val_dict
```

### Image Feature Extraction

```python
def generate_img_features_resnet50(img_paths_dict):
    base_model = ResNet50(weights='imagenet', include_top=True)
    model = Model(inputs=base_model.input, outputs=base_model.get_layer('avg_pool').output)
    img_feature_dict={}
    for img_name in img_paths_dict:
        print(img_name)
        img = image.load_img(train_img_paths[img_name], target_size=(224, 224))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        features = model.predict(x)
        img_feature_dict[img_name]=features[0]
    return img_feature_dict
```

```python
train_img_paths = get_image_paths(train_dict,"train")
val_img_paths = get_image_paths(val_dict,"val")
```

```python
train_img_feature_dict = generate_img_features_resnet50(train_img_paths)
```

```python
f = open("train_img_feature_dict.pkl","wb")
pickle.dump(train_img_feature_dict,f)
f.close()
```

### Data preparation for classification

```python
def get_label_type_distribution(data_dict):
    answer_types={}
    for data in data_dict:
        for label in data["label"]:
            label = label.lower()
            if label in answer_types:
                answer_types[label]+=1
            else:
                answer_types[label]=1
    return answer_types
```

```python
label_distribution = get_label_type_distribution(data_dict["train"])
label_distribution = sorted(label_distribution.items(), key=lambda item: item[1], reverse=True)
```

```python
k = 1000
s = sum(i for j,i in label_distribution[:k])
print("For Top %d labels: Coverage = %.2f"%(k,s/len(train_dict)))
print("Total unique labels: %d"%len(label_distribution))
print("Total labels: %d"%len(train_dict))
```

```python
answer_to_label={}
label_to_answer={}
label=0
for ans,_ in label_distribution[:k]:
    answer_to_label[ans]=label
    label_to_answer[label]=ans
    label+=1
```
